## **What is hierarchical clustering?**


Hierarchical clustering is a hard clustering technique that aims to build a tree-like hierarchy of clusters within the data. This tree-like hierarchical representation of the clusters is called a dendrogram. You can see a dendrogram in the right chart of the figure above.

In general, there are two types of hierarchical clustering approaches:

Bottom-up approach: Also called agglomerative clustering, this approach starts with the individual observations as stand-alone clusters. Then,

It combines the two most similar clusters into one, hence decreasing the total number of clusters by one.
It repeats the previous step above until only a single cluster remains. The figure above illustrates this approach.
Top-down approach: Also called divisive clustering, this approach starts with a single cluster. Then,

It uses a parametric clustering algorithm like k-means to divide the cluster into two clusters.
It repeats the step above for every subcluster until all observations are clustered as stand-alone clusters.

La agrupación jerárquica es una técnica de agrupación estricta que tiene como objetivo construir una jerarquía de agrupaciones en forma de árbol dentro de los datos. Esta representación jerárquica en forma de árbol de los grupos se llama dendrograma. Puede ver un dendrograma en el gráfico derecho de la figura anterior.

En general, existen dos tipos de enfoques de agrupación jerárquica:

Enfoque ascendente: también llamado agrupamiento aglomerativo, este enfoque comienza con las observaciones individuales como grupos independientes. Entonces,

Combina los dos grupos más similares en uno, disminuyendo así el número total de grupos en uno.
Repite el paso anterior hasta que solo quede un grupo. La figura anterior ilustra este enfoque.
Enfoque de arriba hacia abajo: también llamado agrupación divisiva, este enfoque comienza con un solo grupo. Entonces,

Utiliza un algoritmo de agrupamiento paramétrico como k-means para dividir el grupo en dos grupos.
Repite el paso anterior para cada subgrupo hasta que todas las observaciones estén agrupadas como grupos independientes.

Además de la métrica que utilizamos para medir la similitud entre grupos, también debemos decidir cómo seleccionar los elementos representativos de los grupos al calcular las similitudes. Para ello presentamos tres técnicas diferentes que se denominan métodos de vinculación:

## **Linkage methods**
**Ward:** The Ward method tends to look for spherical clusters such that the clusters are cohesive inside and differentiated from other groups. In general, this method tends to find clusters of similar size. Note that the Ward method only works with Euclidean distance.

**Complete:** The Complete method links clusters using their furthest observations, that is, their most dissimilar data points. Consequently, clusters created using this method tend to be comprised of highly similar observations, making the resulting groups quite compact.


**Average:** The Average method links clusters using their centroids. Since this method focuses on centroids and ignoring the boundary points, it tends to create larger groups than the complete method. Moreover, the clusters can be different sizes and shapes in contrast to the Ward method.

## **Métodos de vinculación**
**Ward:** El método Ward tiende a buscar clusters esféricos para que los clusters estén cohesionados en su interior y se diferencien de otros grupos. En general, este método tiende a encontrar grupos de tamaño similar. Tenga en cuenta que el método Ward sólo funciona con distancia euclidiana.

**Completo:** El método Completo vincula grupos utilizando sus observaciones más distantes, es decir, sus puntos de datos más diferentes. En consecuencia, los grupos creados con este método tienden a estar compuestos de observaciones muy similares, lo que hace que los grupos resultantes sean bastante compactos.


**Promedio:** El método Promedio vincula grupos usando sus centroides. Dado que este método se centra en los centroides e ignora los puntos límite, tiende a crear grupos más grandes que el método completo. Además, los grupos pueden tener diferentes tamaños y formas, a diferencia del método Ward.

## **Speed**
The run-time complexity of divisive clustering depends on the clustering technique used in the first step. Since we focused on agglomerative clustering in this checkpoint, let's analyze its complexity in more detail. In agglomerative clustering, we perform n iterations and in each iteration we need to calculate (or update) the similarities between the n clusters. So updating similarities takes n2operations and we do this for each iteration, which in turn makes the time complexity of the agglomerative clustering O(n3).If the number of observations is too high, then the time to run agglomerative clustering will be too long to be practical.

When we have too many observations, we can consider using PCA to reduce dimensionality before applying hierarchical clustering.

## **Velocidad**
La complejidad en tiempo de ejecución del clustering divisivo depende de la técnica de clustering utilizada en el primer paso. Dado que en este punto de control nos centramos en la agrupación aglomerativa, analicemos su complejidad con más detalle. En la agrupación aglomerativa, realizamos n iteraciones y en cada iteración necesitamos calcular (o actualizar) las similitudes entre las n agrupaciones. Entonces, actualizar las similitudes requiere n2 operaciones y hacemos esto para cada iteración, lo que a su vez hace que la complejidad temporal del agrupamiento aglomerativo sea O (n3). Si el número de observaciones es demasiado alto, entonces el tiempo para ejecutar el agrupamiento aglomerativo será demasiado largo para Sé práctico.

## **Agrupación aglomerativa en acción**
Ahora es el momento de implementar agrupaciones aglomerativas. Lo demostraremos en el conjunto de datos de Iris. Aunque puede encontrar el conjunto de datos en Internet, lo cargaremos utilizando el módulo de conjuntos de datos de scikit-learn.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn import datasets, metrics
```

After loading the dataset, we standardize our features as most of the clustering algorithms are sensitive to the scales of the variables.

Note: Clustering algorithms are sensitive to the variable scale of the metrics that are used to analyze similarity. For example, the L2 norm (Euclidean distance), which is a common measure of similarity, is sensitive to variable scale. Similarly, the cosine distance that we'll use in the following implementation of the agglomerative clustering is also sensitive to the mean and hence the scale of the variables. For more information, see this post on Stackoverflow.

Después de cargar el conjunto de datos, estandarizamos nuestras características ya que la mayoría de los algoritmos de agrupamiento son sensibles a las escalas de las variables.

Nota: Los algoritmos de agrupamiento son sensibles a la escala variable de las métricas que se utilizan para analizar la similitud. Por ejemplo, la norma L2 (distancia euclidiana), que es una medida común de similitud, es sensible a la escala variable. De manera similar, la distancia del coseno que usaremos en la siguiente implementación del agrupamiento aglomerativo también es sensible a la media y, por tanto, a la escala de las variables. Para obtener más información, consulte esta publicación en Stackoverflow.

```python
# Loading the data from Sklearn's datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Standarizing the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

We use the AgglomerativeClustering class from scikit-learn's clustering module. We imported it before as from sklearn.cluster import AgglomerativeClustering. Note that, in scikit-learn's jargon, the affinity parameter for AgglomerativeClustering refers to the similarity metric we discussed before. Moreover, the n_clusters parameter denotes the number of clusters we want, which is equivalent to cutting the dendrogram from the respective height. **But don't get confused as agglomerative clustering doesn't require you to specify the number of clusters beforehand. Scikit-learn's implementation uses this parameter just to determine when to stop. So, scikit-learn's implementation of the agglomerative clustering will stop and return the clusters when the specified n_clusters value is reached.

We use the complete method for linkage and cosine for measuring the similarities of the clusters. We also set the n_clusters=3 as the Iris dataset includes three species. You can play with the other methods and values and compare their results with each other.

Usamos la clase AgglomerativeClustering del módulo de agrupación en clústeres de scikit-learn. Lo importamos antes desde sklearn.cluster import AgglomerativeClustering. Tenga en cuenta que, en la jerga de scikit-learn, el parámetro de afinidad para AgglomerativeClustering se refiere a la métrica de similitud que discutimos antes. Además, el parámetro n_clusters denota el número de clusters que queremos, lo que equivale a cortar el dendrograma desde la altura respectiva. **Pero no se confunda, ya que la agrupación aglomerativa no requiere que especifique la cantidad de grupos de antemano. La implementación de Scikit-learn utiliza este parámetro solo para determinar cuándo detenerse. Por lo tanto, la implementación de scikit-learn del clustering aglomerativo se detendrá y devolverá los clusters cuando se alcance el valor n_clusters especificado.

Utilizamos el método completo de enlace y coseno para medir las similitudes de los grupos. También configuramos n_clusters=3 ya que el conjunto de datos de Iris incluye tres especies. Puedes jugar con los otros métodos y valores y comparar sus resultados entre sí.

```python
# Defining the agglomerative clustering
agg_cluster = AgglomerativeClustering(linkage='complete',
                                      affinity='cosine',
                                      n_clusters=3)

# Fit model
clusters = agg_cluster.fit_predict(X_std)

```

Below, we sketch the clusters assigned by agglomerative clustering. To be able to do that, we apply PCA to the four features we have and use the first two principal components.

In the plot below, the numbers show the cluster each observation has been assigned to by the algorithm. The colors denote the true classes.


A continuación, esbozamos los conglomerados asignados por agrupamiento aglomerativo. Para poder hacer eso, aplicamos PCA a las cuatro características que tenemos y usamos los dos primeros componentes principales.

En el siguiente gráfico, los números muestran el grupo al que el algoritmo ha asignado cada observación. Los colores denotan la verdadera clase.

```python
pca = PCA(n_components=2).fit_transform(X_std)

plt.figure(figsize=(10,5))
colours = 'rbg'
for i in range(pca.shape[0]):
    plt.text(pca[i, 0], pca[i, 1], str(clusters[i]),
             color=colours[y[i]],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

Let's get the Adjusted Rand Index (ARI) and the silhouette scores for the agglomerative clustering approach and compare them with the results for k-means from the previous checkpoint:

Obtengamos el índice de Rand ajustado (ARI) y las puntuaciones de silueta para el enfoque de agrupamiento aglomerativo y compárelos con los resultados de k-medias del punto de control anterior:

```python
print("Adjusted Rand Index of the Agglomerative Clustering solution: {}"
      .format(metrics.adjusted_rand_score(y, clusters)))
print("The silhoutte score of the Agglomerative Clustering solution: {}"
      .format(metrics.silhouette_score(X_std, clusters, metric='euclidean')))
```

The ARI score for the agglomerative clustering approach is 0.63 and slightly higher than the one for k-means, which was 0.62. However, the silhouette score of the algorithm is 0.45, which is slightly lower than that of the k-means which was 0.46. Overall, the comparison is not very decisive and both algorithms achieve more or less the same performance.
